```python
from tqdm.notebook import tqdm

import analysis
import catalog
import energy
import histograms

catalog.display_params_for_catalog()
tqdm.pandas()
```

```python
ESEC_avalanches = catalog.load("catalog/7_ESEC_avalanches_42.pickle")

histograms.open_catalog(ESEC_avalanches, pos_number=0.15, xlim=50)
```

```python
for event_index in tqdm(ESEC_avalanches["numero"], total=len(ESEC_avalanches)):
    print("-------------------------")
    print("Event numero", event_index)
    print("-------------------------")
    
    event, stream, trace = analysis.filter_stream(ESEC_avalanches, event_index, 0, freq_HP=9, freq_LP=0.5)

    #garder seulement la première station
    if len(stream) > 2:
        stream = stream[:1]
        print("Stream size :", len(stream))
   
    energy.compute(ESEC_avalanches, stream, event_index)
```
