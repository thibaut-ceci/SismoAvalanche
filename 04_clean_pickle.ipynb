# Remove useless pickle

Here, we want to remove events : 
- without digital data
- without pickle file

And stations :
- with a frequency below of 20 Hz in pickle file
- without instrumental response in pickle file

```python
import catalog
import cleaning

catalog.display_parameters()
```

```python
ESEC_avalanches = catalog.load("catalog/4_ESEC_avalanches_88.pickle")

catalog.open_plot(ESEC_avalanches, pos_number=0.11, xlim=100)
```

## Removing events without digital data

```python
ESEC_avalanches = cleaning.events_without_digital_data(ESEC_avalanches)
```

## Removing events without pickle file

No data was downloaded probably because no stations were found between the source and 5 degrees or IRIS does not offer data for this event

```python
ESEC_avalanches = cleaning.events_without_pickle_file(ESEC_avalanches)
```

## Here, we create a new loop to remove pickle files with a frequency below of 20 Hz

```python
ESEC_avalanches = cleaning.trace_with_a_frequency_below_of(20, ESEC_avalanches)
```

## Here, we create a new loop to remove pickle files without instrumental response

```python
cleaning.trace_without_metadata(ESEC_avalanches)
```

```python
catalog.open_plot(ESEC_avalanches, 0.11, 90)

ESEC_avalanches.to_pickle("catalog/5_ESEC_avalanches_83.pickle")
```
