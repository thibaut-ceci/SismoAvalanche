# Map events

In this notebook, we want to see the events in the catalog with their stations.

```python
from tqdm.notebook import tqdm

import catalog
import figures
import histograms
import mapper

catalog.display_params_for_catalog()
tqdm.pandas()
```

## Load the catalog

```python
ESEC_avalanches = catalog.load("catalog/6_ESEC_avalanches_83.pickle")

histograms.open_catalog(ESEC_avalanches, pos_number=0.15, xlim=90)
```

## See all the events in a world map

```python
World, Europe, Japon, Groenland, France, Amerique_du_nord, Iles_Mariana, Suisse, Chine, World_without_Antarctica, manuel, manuel2 = mapper.select_extent()

mapper.show_select_area(ESEC_avalanches, World, add_equator_and_tropics_on_map = False, add_legend = False)

figures.save("figures/mapper/world_map_events.pdf", tight_layout=False)
```

## See one event in the catalog

```python
#Choose an event :
event_index = 11

mapper.show(ESEC_avalanches.iloc[event_index], max_extent=5)

figures.save(f"figures/mapper/map_{event_index}.pdf", tight_layout=False)
```

## Get the most instrumented avalanches

```python
most_instrumented_avalanches = catalog.get_the_most_or_least_instrumented_avalanches(ESEC_avalanches, 3, ascending=False)

#Plot the most instrumented events
for event_index, event in most_instrumented_avalanches.iterrows():
    print("Event number :", event_index)
    mapper.show(event, max_extent=5)
    figures.save(f"figures/mapper/map_{event_index}.pdf", tight_layout=False)
```

## Get the least instrumented avalanches

```python
least_instrumented_avalanches = catalog.get_the_most_or_least_instrumented_avalanches(ESEC_avalanches, 3, ascending=True)

#Plot the least instrumented events
for event_index, event in least_instrumented_avalanches.iterrows():
    print("Event number :", event_index)
    mapper.show(event, max_extent=11)
    figures.save(f"figures/mapper/map_{event_index}.pdf", tight_layout=False)
```
