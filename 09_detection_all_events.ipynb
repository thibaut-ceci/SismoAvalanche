```python
from tqdm.notebook import tqdm

import matplotlib.pyplot as plt

import analysis
import catalog
import figures

tqdm.pandas()
catalog.display_parameters()
```

## Reading catalog

```python
ESEC_avalanches = catalog.load("catalog/6_ESEC_avalanches_83.pickle")
 
catalog.open_plot(ESEC_avalanches, pos_number=0.11, xlim=90)
```

```python
plt.figure(figsize=(10, 15), dpi=300)

events_to_keep = []



for event_index in tqdm(ESEC_avalanches["numero"], total=len(ESEC_avalanches)):
    print("-------------------------")
    print("Event numero", event_index)
    print("-------------------------")

    _, stream, _ = analysis.filter_stream(ESEC_avalanches, event_index, 0, freq_HP=9, freq_LP=0.5)

    print("Nombre de traces :", len(stream))

    analysis.detection_method_for_all_events(ESEC_avalanches, stream, event_index, events_to_keep, label1 = True, label2 = True)

plt.xlabel(r"Volume ($\mathrm{m^3}$)", fontsize=14)
plt.ylabel("Distance de la station  l'avalanche (km)", fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.xscale("log")
plt.ylim(0, 600)
plt.margins(y=0)
# plt.legend(loc="lower right", fontsize=14)
figures.save(f"figures/all_events")
plt.show()
```

```python
ESEC_avalanches = ESEC_avalanches.loc[events_to_keep]

catalog.open_plot(ESEC_avalanches, pos_number=0.11, xlim=50)
```
