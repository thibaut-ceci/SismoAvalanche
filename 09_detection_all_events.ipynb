# Detection on all events (in the same fig)

Here, we want to test the detection method in all stations in all events. We will create a loop to see the detection method in all streams.

```python
from tqdm.notebook import tqdm

import matplotlib.pyplot as plt

import analysis
import catalog
import figures

tqdm.pandas()
catalog.display_parameters()
```

## Load catalog

```python
ESEC_avalanches = catalog.load("catalog/6_ESEC_avalanches_83.pickle")

## In the code 08_detection_all_events_stream, it is possible to notice several strange detections that will be removed now so as not to alter the results later.
## This events will be removed because the seismic signal is damaged : 
events_to_remove = [7, 13, 16, 31, 34, 54, 58, 59, 61, 76]
ESEC_avalanches = ESEC_avalanches.drop(events_to_remove, axis=0)

## This events will be remove because the detection is too bad : 
events_to_remove = [1, 15, 19, 26, 29, 33, 35, 44, 56, 64, 65, 71]
ESEC_avalanches = ESEC_avalanches.drop(events_to_remove, axis=0)

catalog.open_plot(ESEC_avalanches, pos_number=0.11, xlim=70)
```

## The detection method in all events

```python
plt.figure(figsize=(10, 15), dpi=300)

## Initialize a list to keep the detected events
events_to_keep = []

## Loop to iterate over all the events in ESEC
for event_index in tqdm(ESEC_avalanches["numero"], total=len(ESEC_avalanches)):

    print("-------------------------")
    print("Event number", event_index)
    print("-------------------------")

    ## Filter the stream
    _, stream, _ = analysis.filter_stream(ESEC_avalanches, event_index, 0, freq_HP=9, freq_LP=0.5)

    print("Number of traces :", len(stream))

    ## The detection method
    analysis.detection_method_for_all_events(ESEC_avalanches, stream, event_index, events_to_keep, label1 = True, label2 = True)

plt.xlabel(r"Volume ($\mathrm{m^3}$)", fontsize=14)
plt.ylabel("Distance from station to avalanche [km]", fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=12)
plt.xscale("log")
plt.ylim(0, 600)
plt.margins(y=0)

figures.save(f"figures/all_events")

plt.show()
```

## Keep only events with a detection in the first trace

```python
## Thanks to the detection method, it is possible to delete events that do not have possible detection on their first trace
ESEC_avalanches = ESEC_avalanches.loc[events_to_keep]

catalog.open_plot(ESEC_avalanches, pos_number=0.11, xlim=50)

ESEC_avalanches.to_pickle("catalog/7_ESEC_avalanches_44.pickle")
```
